```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
datasets = ['Dataset-1', 'Dataset-2', 'Dataset-3', 'Dataset-4', 'Dataset-5']

# AR mining tools 的时间
ar_mining_time_1 = [0.0099, 0.0061, 0.0295, 0.0097, 0.2870]  # min_support = 0.1, min_confidence = 0.5
ar_mining_time_2 = [0.2163, 0.0143, 0.0272, 0.2491, 0.0091]  # min_support = 0.4, min_confidence = 0.7

# Brute-force 的时间
brute_force_time_1 = [283.2342, 57.1152, 328.6313, 322.9166, 320.3835]  # min_support = 0.1, min_confidence = 0.5
brute_force_time_2 = [277.4457, 57.3230, 322.9803, 322.9933, 325.1465]  # min_support = 0.4, min_confidence = 0.7

# 设置柱状图的宽度
bar_width = 0.2
x = np.arange(len(datasets))

# 创建图表
plt.figure(figsize=(12, 6))

# 绘制 AR mining tools 的柱状图
plt.bar(x - bar_width, ar_mining_time_1, width=bar_width, label='AR Mining (Support=0.1, Confidence=0.5)', color='blue')
plt.bar(x, ar_mining_time_2, width=bar_width, label='AR Mining (Support=0.4, Confidence=0.7)', color='cyan')

# 绘制 Brute-force 的柱状图
plt.bar(x + bar_width, brute_force_time_1, width=bar_width, label='Brute-force (Support=0.1, Confidence=0.5)', color='red')
plt.bar(x + 2 * bar_width, brute_force_time_2, width=bar_width, label='Brute-force (Support=0.4, Confidence=0.7)', color='orange')

# 添加标签和标题
plt.xlabel('Datasets')
plt.ylabel('Time (s)')
plt.title('Comparison of Measured and Estimated Times')
plt.xticks(x + bar_width / 2, datasets)
plt.yscale('log')  # 使用对数刻度方便比较数值差异较大的数据

# 添加图例
plt.legend()

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据集标签
datasets = ['Dataset-1', 'Dataset-2', 'Dataset-3', 'Dataset-4', 'Dataset-5']

# AR mining tools 的时间
ar_mining_time_1 = [0.0172, 0.1104, 0.3313, 0.0169, 0.0182]  # min_support = 0.1, min_confidence = 0.5
ar_mining_time_2 = [0.0154, 0.0151, 0.0137, 0.0192, 0.1531]  # min_support = 0.4, min_confidence = 0.7

# 设置柱状图的宽度
bar_width = 0.3
x = np.arange(len(datasets))

# 创建图表
plt.figure(figsize=(10, 5))

# 绘制 AR mining tools 的柱状图
plt.bar(x - bar_width/2, ar_mining_time_1, width=bar_width, label='AR Mining (Support=0.1, Confidence=0.5)', color='blue')
plt.bar(x + bar_width/2, ar_mining_time_2, width=bar_width, label='AR Mining (Support=0.4, Confidence=0.7)', color='cyan')

# 标记 Brute-force 数据无法获得的情况
for i, dataset in enumerate(datasets):
    plt.text(i, max(ar_mining_time_1 + ar_mining_time_2) * 1.1, 'NA', ha='center', color='red', fontsize=10)

# 添加标签和标题
plt.xlabel('Datasets')
plt.ylabel('Time (s)')
plt.title('AR Mining Tools vs Brute-force (Brute-force Unavailable for ItemCount=200)')
plt.xticks(x, datasets)

# 添加图例
plt.legend()

# 显示图表
plt.show()
```
